```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
x=[1,2,3,4,5]
y=[1,1,2,2,4]
plt.scatter(x,y)
plt.xlabel("advertising cost")
plt.ylabel("sales")
plt.show()
import matplotlib.pyplot as plt
from scipy import stats
x=[1,2,3,4,5]
y=[0.5,1.5,2.0,2.5,3.5]
slope,intercept,r,p,std_err=stats.linregress(x,y)
def myfunc(x):
    return slope*+intercept
mymodel=list(map(myfunc,x))
print("slope:",slope)
print("intercept:",intercept)
print("predicted values:",mymodel)
print("standart error:",std_err)
print("correlation coefficient(r):",r)
plt.scatter(x,y)
plt.xlabel("advertising cost")
plt.ylabel("sales")
plt.plot(x,mymodel,lw=4,c='orange')
plt.show()
```
